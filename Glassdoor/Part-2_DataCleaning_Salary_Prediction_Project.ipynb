## Column Formating

```python
### Column name : Type Column ###

def type_format(x):
    
    ''' Here we will convert various sectors into either 
    a "Private" Or a "Public" sector '''
    
    if "private" in x.lower():
        return "Private"
    elif "public" in x.lower():
        return "Public"
    elif x in ["Subsidiary or Business Segment","College / University","Franchise","Contract"]:
        return "Private"
    elif x in ["Non-profit Organisation","Government"]:
        return "Public"
    else:
        return x

## Call function on "Type" column
df["Type"]=df["Type"].apply(type_format)

## Check if replacement done properly
df["Type"].unique()
```

```python
### Column Name : Industry ###

def Common_industry(x):
    
    ''' There Are Multiple Similar Industries, Which Can Be Combined Together
    Under One Common Industry Name To Reduce The Number Of Unique Values In 
    Industry Column '''
    
    Banking_Finance=["Banks & Building Societies","Investment Banking & Asset Management",
                     "Lending","Accounting","Financial Analytics & Research","Insurance Operators",
                     "Financial Transaction Processing"]

    Consulting=["Consulting","Publishing"]

    Education=["Education Training Services","Colleges & Universities"]

    GovernmentAgencies=["Government Agencies","Energy","Oil & Gas Services","Aerospace & Defence",
                        "Research & Development","Social Services"]

    HealthcareMedicine=["Healthcare Services & Hospitals","Biotech & Pharmaceuticals"]

    InformationTechnology=["IT Services","Enterprise Software & Network Solutions","Computer Hardware & Software","Internet"]

    Manufacturing=["Industrial Manufacturing","Electrical & Electronic Manufacturing","Food & Drink Manufacturing",
                   "Consumer Products Manufacturing","Transportation Equipment Manufacturing","Department",
                   "Clothing, & Shoe Shops","Architectural & Engineering Services"]

    Outsource="Staffing & Outsourcing","Advertising & Marketing"

    RealEstate=["Real Estate","Brokerage Services"]

    Telecommunication_Television=["Telecommunications Services","Film Production & Distribution Cable", 
                                  "Internet & Telephone Providers","TV Broadcasting & Cable Networks"]

    Transportation=["Transportation Management","Bus & Coach Services","Shipping","Logistics & Supply Chain"]
                   
    x=x.strip()
                   
    if x in Banking_Finance:
        return "Banking/Finance"
    elif x in Consulting:
        return "Consulting"
    elif x in Education:
        return "Education"
    elif x in GovernmentAgencies:
        return "Government Agencies"
    elif x in HealthcareMedicine:
        return "Healthcare Medicine"
    elif x in InformationTechnology:
        return "Information Technology"
    elif x in Manufacturing:
        return "Manufacturing"
    elif x in Outsource:
        return "Outsource"
    elif x in RealEstate:
        return "Real Estate"
    elif x in Telecommunication_Television:
        return "Telecommunication/Television"
    elif x in Transportation:
        return "Transportation"
    else:
        return x  

## call function on "Industry" column
df["Industry"]=df["Industry"].apply(Common_industry)

## verify combined industries
df["Industry"].unique()
```

```python
### Column Name : Basic_Pay ###

def format_Basic_Pay(x):
    
    ''' Remove "₹" Sign and Comma's From Currancy Values,
    also replace invalid entries with Unknown '''
    
    if x=='No salaries reported.' or x=="NA":
        return "Unknown"
    elif "-" in x:
        a=x.split("-")[0].strip()
        a=a.replace('₹', '').replace(',', '')
        b=x.split("-")[1].strip()
        b=b.replace('₹', '').replace(',', '')
        return (int(a)+int(b))/2
    else:
        c=x.strip().replace('₹', '').replace(',', '')
        return int(c)
    
df["Basic_Pay"]=df["Basic_Pay"].apply(format_Basic_Pay)
```

```python
### Column Name : Avg_Pay ###

def format_Avg_Pay(x):
    
    ''' Remove "₹" Sign And Comma's From Avg_Pay values,
    also replace invalid values with Unknown'''
    
    if x=='No salaries reported.' or x==np.nan:
        return "Unknown"
    if type(x)==str:
        c=x.strip().replace('₹', '').replace(',', '')
        return int(c)
    else:
        return x
df["Avg_Pay"]=df["Avg_Pay"].apply(format_Avg_Pay)
```

```python
### Column Name: Avg_Pay_Num ###

df.Avg_Pay_Num.fillna("Unknown",inplace=True)

def format_Avg_Pay_Num(x):
    
    ''' Remove "₹" Sign And Comma's from Avg_Pay_Num column'''
    
    if x=="Unknown":
        return "Unknown"
    elif "-" in x:
        a=x.split("-")[0].strip()
        a=a.replace('₹', '').replace(',', '')
        b=x.split("-")[1].strip()
        b=b.replace('₹', '').replace(',', '')
        return (int(a)+int(b))/2
    else:
        c=x.strip().replace('₹', '').replace(',', '')
        return int(c)
    
df["Avg_Pay_Num"]=df["Avg_Pay_Num"].apply(format_Avg_Pay_Num)
```

```python
### Column Name : Pay Avg_Pay & Avg_Pay_Num ###

df["Avg_Pay"].fillna("Unknown",inplace=True)

def combine_Avg_Pay_Avg_Pay_Num(Avg_Pay,Avg_Pay_Num):
    
    ''' fill the values from Avg_Pay_Num into Avg_Pay, where
    Avg_Pay have Unknown values '''
    
    if Avg_Pay=="Unknown":
        return Avg_Pay_Num
    else:
        return Avg_Pay
    
df["Avg_Pay"]=df[["Avg_Pay","Avg_Pay_Num"]].apply(lambda row: combine_Avg_Pay_Avg_Pay_Num(row["Avg_Pay"],row["Avg_Pay_Num"]),axis=1)
```

### Linear Regression For Null Value Prediction

```python
### Apply a linear regression to fill the null values of Founded column with respect to Size ###

df_Founded=df[["Size","Founded"]]

df_Founded["Size"].replace({'1001 to 5000 Employees':4, '501 to 1000 Employees':3,
       '10000+ Employees':6, '5001 to 10000 Employees':5,
       '51 to 200 Employees':1, '1 to 50 Employees':0, '201 to 500 Employees':2},inplace=True)

train=df_Founded.dropna(axis=0)

x=train.drop("Founded",axis=1)
y=train["Founded"]

from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x,y)

print("1 to 50 Employees",model.predict([[0]]))
print("51 to 200 Employees",model.predict([[1]]))
print("201 to 500 Employees",model.predict([[2]]))
print("501 to 1000 Employees",model.predict([[3]]))
print("1001 to 5000 Employees",model.predict([[4]]))
print("5001 to 10000 Employees",model.predict([[5]]))
print("10000+ Employees",model.predict([[6]]))

df["Founded"].fillna("Unknown",inplace=True)

import matplotlib.pyplot as plt
founded_hist=[]
for i in df.Founded.unique():
    if i!="Unknown":
        founded_hist.append(i)        
plt.hist(founded_hist,bins=10)
```

```python
### Fill the prediction of null values into Founded column and check the distrubution ###

def founded(x,y):
    
    ''' fill the null values of Founded using values predicted by above linear regression'''
    
    a={"1 to 50 Employees":2020,"51 to 200 Employees":2015,"201 to 500 Employees":2003,"501 to 1000 Employees":1991,
                   "1001 to 5000 Employees":1979,"5001 to 10000 Employees":1968,"10000+ Employees":1956}
    if y=="Unknown":
        return a[x]
    else:
        return y
    
df["Founded"]=df[["Size","Founded"]].apply(lambda row:founded(row["Size"],row["Founded"]),axis=1)

import matplotlib.pyplot as plt
plt.hist(df.Founded,bins=20)
```

```python
### For Type column, fill the Unknown valus with column mode ###

df.Type.replace({"Subsidiary or Business Segment":"Private","College / University":"Public","Non-profit Organisation":"Public",
                "Government":"Public","Contract":"Private"},inplace=True)

df.Type=df.Type.apply(lambda x: "Private" if "private" in x.lower() else x)
df.Type=df.Type.apply(lambda x: "Public" if "public" in x.lower() else x)

print("mode =>",df.Type.mode()[0],end="\n\n")
print(df.Type.value_counts())

df.Type.replace("Unknown","Private",inplace=True)
```

```python
### Fill unknown values of Industry with column mode ###

print(df.Industry.mode()[0])
df.Industry.replace("Unknown / Non-Applicable","Information Technology",inplace=True)
```

### SVM to Predict Null Values

```python
### Apply SVM to fill the null values of Revenue columns ###

df.Revenue.replace("Unknown / Non-Applicable","Unknown",inplace=True)
df_revenue=df[["Size","Founded","Type","Revenue"]]

df_revenue["Revenue"].replace({"$10+ billion (USD)":11,"$5 to $10 billion (USD)":10,"$2 to $5 billion (USD)":9,"$1 to $2 billion (USD)":8,
"$500 million to $1 billion (USD)":7,"$100 to $500 million (USD)":6,"$50 to $100 million (USD)":5,
"$25 to $50 million (USD)":4,"$10 to $25 million (USD)":3,"$5 to $10 million (USD)":2,"$1 to $5 million (USD)":1,
"Less than $1 million (USD)":0},inplace=True)
df_revenue["Size"].replace({"1 to 50 Employees":0,"51 to 200 Employees":1,"201 to 500 Employees":2,"501 to 1000 Employees":3,
                   "1001 to 5000 Employees":4,"5001 to 10000 Employees":5,"10000+ Employees":6},inplace=True)
df_revenue["Type"].replace({"Private":1,"Public":0},inplace=True)

# Training
x=df_revenue[df_revenue["Revenue"]!="Unknown"].drop("Revenue",axis=1)
y=df_revenue[df_revenue["Revenue"]!="Unknown"]["Revenue"]
y=y.astype('int')

# Testing
x1=df_revenue[df_revenue["Revenue"]=="Unknown"].drop("Revenue",axis=1)
y1=df_revenue[df_revenue["Revenue"]=="Unknown"]["Revenue"]
```

```python
### Remove the Unwanted Columns ###

df.Job_Title="Data Scientist"
df.drop(["Additional_Pay","Avg_Pay_Num","Avg_Pay_Text","Max_Pay","Min_Pay"],axis=1,inplace=True)
```

```python
### Formating Positive,Negative,Neutral,Online,Campus,Refferal and Rating all together ###

## Remove "%" sign from columns
for i in ["Positive","Negative","Neutral","Online","Campus","Refferal"]:
    df[i]=df[i].str[:-1]

    
## Fill null values with column mean 
for i in ["Rating","Positive","Negative","Neutral","Online","Campus","Refferal"]:
    df[i]=df[i].apply(pd.to_numeric)
    df[i].fillna(df[i].mean(),inplace=True)
    
df.isnull().sum()
```

```python
### Convert the montly Avg_Pay and Basic_Pay into annual package ###

## First of all lets make a new column stating if salaries are annualy or monthly
df["Monthly"]=df["Job_Title"].apply(lambda x: "Yes" if "monthly" in x.lower() else "No")
print(df.Monthly.value_counts())

def monthtoannual(x,y):
    
    ''' Where Salary Is Monthly Converting It Into Annual Package'''
    
    if x=="Yes":
        return y*12
    else:
        return y
    
df.Basic_Pay=df[["Monthly","Basic_Pay"]].apply(lambda row:monthtoannual(row["Monthly"],row["Basic_Pay"]),axis=1)

df.Avg_Pay=df[["Monthly","Avg_Pay"]].apply(lambda row:monthtoannual(row["Monthly"],row["Avg_Pay"]),axis=1)
```

```python
### Fill the Unknown values of Basic_Pay and Avg_Pay with column mean ###

df.Basic_Pay.replace("Unknown",round(df[df["Basic_Pay"]!="Unknown"]["Basic_Pay"].mean(),2),inplace=True)
df.Avg_Pay.replace("Unknown",round(df[df["Avg_Pay"]!="Unknown"]["Avg_Pay"].mean(),2),inplace=True)
```

```python
### Column Name : Headquarter ###

## Clean the Headquarter
df["Headquarter"]=df["Headquarter"].apply(lambda row: row.split(",")[1])
df["Headquarter"]=df["Headquarter"].str.strip()

conuntries=[]
for i in df["Headquarter"].unique():
    if len(i)==2:
        conuntries.append(i)
        
full_name=["New York","Morocco","Michigan","Connecticut","New Joursey","California","Washington","Illinois","Ohio","Missouri","Minnesota",
"Alabama","Pennsylvania","Texas","Hanover","North Carolina","South Carolina","Florida","Virginia","Georgia","Nevada","Arizona",
"Wisconsin","Kansas","Colorado","Oklahoma","Louisiana"]

df["Headquarter"].replace(dict(zip(conuntries,full_name)),inplace=True)
df["Headquarter"].unique()
```
